```python
import argparse
import pdb

from pyshex.shex_evaluator import evaluate_cli as shexeval
from pyshex import shex_evaluator
from sparqlslurper import SlurpyGraph
from pyshex.user_agent import SlurpyGraphWithAgent
```

```python
permagraph = None
OUTPUT_FORMATS = ["ttl", "rdf/xml", "rdf", "n3", "nquads"]
```

```python
def setup_persistent_slurper():
    def persistent_slurper(rdf: str, *args, **kwargs) -> SlurpyGraph:
        global permagraph
        permagraph = SlurpyGraphWithAgent(rdf, *args, **kwargs)
        return permagraph
    shex_evaluator.SlurpyGraphWithAgent = persistent_slurper
```

```python
def slurpWikidata(offset = 0):
    
    offset = offset * 200
    
    setup_persistent_slurper()

    sparql_query = """
    PREFIX wdt: <http://www.wikidata.org/prop/direct/>
    PREFIX wd: <http://www.wikidata.org/entity/>

    SELECT DISTINCT ?item WHERE { ?item wdt:P2063 ?lipidMapsId. } ORDER BY ?item LIMIT 200 offset 
    
    """   
    sparql_query = sparql_query.strip() + " " +str(offset)

    print(sparql_query)
    
    rdf = "https://raw.githubusercontent.com/kg-subsetting/biohackathon2021/main/use_cases/lipidmaps/lipids.shex"

    options = ["-ss",
               "-sq", sparql_query,
               "http://query.wikidata.org/sparql",
               rdf,
               "--stopafter", "0",
               "-ps"]
    shexeval(options)
    serialization = permagraph.serialize(format="ttl").decode()

    text_file = open("wikidata/wikidata-lipids-"+str(offset)+".ttl", "w")

    print("Writting output to file...")
    text_file.write(serialization)
    text_file.close()
```

```python
import os

def slurpJob():    
    try:
        for i in range(110):
            if(not os.path.isfile("wikidata/wikidata-lipids-"+str(i*200)+".ttl")):
                print("Doing offset"+str(i))
                slurpWikidata(i)
    except Exception as e:
        slurpJob()
```
